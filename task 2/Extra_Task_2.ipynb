#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
####################################### Line Plot using Matplotlib visualization #########################################
plt.plot(df.month_number , df.total_profit)

plt.xlim(1 , 12)
plt.ylim(100000 , 500000)

plt.title('Company profit per month')
plt.xlabel('Month number')
plt.ylabel('Total profit')

x_ticks = np.arange(1 , 13 , 1)
y_ticks = np.arange(100000, 500001 , 100000)
plt.xticks(x_ticks)
plt.yticks(y_ticks)

plt.show()
```

```python
####################################### Line Plot using Seaborn visualization #########################################
sns_plt = sns.lineplot(x = df.month_number , y = df.total_profit)
sns_plt.set(
    title = 'Company profit per month',
    xlabel = 'Month number',
    ylabel = 'Total profit',
  
)
sns_plt.set_xlim(1 , 12 , 1)
sns_plt.set_ylim(100000 , 500000 , 100000)


x_ticks = np.arange(1 , 13 , 1)
y_ticks = np.arange(100000, 500001 , 100000)
sns_plt.set_xticks(x_ticks)
sns_plt.set_yticks(y_ticks)

plt.show()
```

```python
#################################### Line Plot using Pandas visualization #####################################
df.plot.line(x='month_number' , y = 'total_profit' ,
             title = 'Company profit per month',
             xlim = (1 , 12),
             ylim = (100000 , 500000),
             xticks = x_ticks ,
             yticks = y_ticks,
             legend = False
            ).plot()
```

#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
df.columns
```

```python
############################## multiline plots using Matplotlib visualization #########################################
plt.plot(df.month_number , df.facecream ,'o-' , linewidth = 3 , label = 'Face Cream Sales Data')
plt.plot(df.month_number , df.facewash  ,'o-' , linewidth = 3 , label = 'Face Wash Sales Data')
plt.plot(df.month_number , df.toothpaste ,'o-' , linewidth = 3 , label = 'Toothpaste Sales Data')
plt.plot(df.month_number , df.bathingsoap ,'o-' , linewidth = 3 , label = 'bathingsoap Sales Data')
plt.plot(df.month_number , df.shampoo,'o-' , linewidth = 3  , label = 'shampoo Sales Data')
plt.plot(df.month_number , df.moisturizer ,'o-' , linewidth = 3 , label = 'moisturizer Sales Data')
# plt.xlim(1 , 12)
# plt.ylim(100000 , 500000)

plt.title('Sales Data')
plt.xlabel('Month number')
plt.ylabel('Sales unites in number')

plt.legend()

x_ticks = np.arange(1 , 13 , 1)
y_ticks = np.arange(2000, 13000 , 2000)
plt.xticks(x_ticks)
plt.yticks([1000] + y_ticks.tolist() + [15000 , 18000])

plt.show()
```

```python
########################### multiline plots Plot using Seaborn visualization #########################################
l = ['facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo', 'moisturizer']
for i in l :
    sns_plt = sns.lineplot(linewidth = 3, marker = 'o', dashes = False , x = df.month_number , y = df[i] , label = i + 'Sales Data')
    
sns_plt.set(
    title = 'Sales Data',
    xlabel = 'Month number',
    ylabel = 'Sales unites in number',
  
)


x_ticks = np.arange(1 , 13 , 1)
y_ticks = np.arange(2000, 13000 , 2000)
sns_plt.set_xticks(x_ticks)
sns_plt.set_yticks([1000] + y_ticks.tolist() + [15000 , 18000])


plt.show()
```

```python
#################################### multiline plots using Pandas visualization #####################################

df.plot.line(x='month_number' , y = l , 
             title = 'Sales Data',
             ylabel = 'Sales unites in number',
             xticks = x_ticks ,
             yticks =[1000] +  y_ticks.tolist() + [15000 , 18000] ,
             style = 'o-'
            ).plot()
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
####################################### histogram Plot using Matplotlib visualization #########################################



plt.xlim(145000 , 360000)

x_ticks = np.arange(150000 , 250001 , 25000)
plt.xticks(x_ticks.tolist() + [300000 , 350000])
y_ticks = np.arange(0 , 5.5 , 1)
plt.yticks(y_ticks)
plt.hist(df.total_profit , bins =x_ticks.tolist() + [300000 , 350000] , label = 'profit data')
plt.legend()
plt.title('profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual profit in dollar')
```

```python
####################################### histogram  Plot using Seaborn visualization #########################################

sns.histplot(data= df,
             x='total_profit' ,
             bins = x_ticks.tolist() + [300000 , 350000] ,
             label = 'profit data',
             edgecolor = None,
             legend = True
            )

plt.xticks(x_ticks.tolist() + [300000 , 350000])
y_ticks = np.arange(0 , 5.5 , 1)
plt.yticks(y_ticks)
plt.title('profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual profit in dollar')

plt.show()
```

```python
#################################### histogram Plot using Pandas visualization #####################################

df.hist(column = 'total_profit' ,
        bins =  x_ticks.tolist() + [300000 , 350000] ,
        grid = False,
        legend = 'profit data'
       )
plt.xticks(x_ticks.tolist() + [300000 , 350000])
y_ticks = np.arange(0 , 5.5 , 1)
plt.yticks(y_ticks)
plt.title('profit data')
plt.xlabel('profit range in dollar')
plt.ylabel('Actual profit in dollar')
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
#################################### pie Plot using Matplotlib visualization #####################################
sum_of_sales = []
for i in l :
    sum_of_sales.append(df[i].sum())

plt.pie(sum_of_sales, labels= l ,autopct="%1.1f%%")
plt.legend(loc = 'lower right')
plt.show() 
```

```python
#################################### pie Plot using Pandas visualization #####################################

df2 = pd.DataFrame(index = l , data = sum_of_sales)
df2.plot.pie(subplots=True , autopct='%1.1f%%')
plt.legend(loc = 'lower right')
```
