# Analyze data from kaggle "Data Science Job Salaries" Data Set

```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('ds_salaries.csv')
```

```python
sns.set(style="darkgrid")
sns.displot(data=df, x='salary_in_usd', col="company_size" , bins=20)
plt.xticks(np.arange(0 , 600000 , 75000))
```

Average salaries for large companies range between    50000(usd)   to   150000 <br>
Average salaries for midum companies range between    50000(usd)   to   200000 <br>
Average salaries for large companies range between    25000(usd)   to   100000 

```python
sns.swarmplot(data=df, y = 'salary_in_usd', x = 'experience_level' , palette="Set2")
plt.title('distribution of salary based on the experience level', fontsize=15, fontweight='bold')
plt.show()
```

```python
sns.set(style="darkgrid")
sns.displot(data=df, x='salary_in_usd', col="experience_level" , bins=10)
```

The previous figure shows the change in the salary rate with a change in the level of experience

```python
sns.set(style="darkgrid")
sns.displot(data=df, x='remote_ratio', col="work_year" , bins=10)
plt.yticks(np.arange(0 , 310 , 20))
plt.xticks(np.arange(0 , 105 , 10))
```

Through this figure, we notice a significant increase in reliance on remote work as time progresses

```python
sns.set(style="darkgrid")
sns.displot(
    data=df,
    x='remote_ratio',
    col="company_size",
    bins=10,
    
)

plt.yticks(np.arange(0 , 310 , 20))
plt.xticks(np.arange(0 , 105 , 10))

plt.show()
```

From this figure, we notice that as the size of the company increases, the reliance on remote work increases.

```python
# sns.countplot(df , x ='experience_level' )

plt.pie(df['experience_level'].value_counts() ,
        labels = df['experience_level'].value_counts().index ,
        autopct='%1.1f%%')
plt.title('distribution of employees based on the experience level', fontsize=15, fontweight='bold')
```

the most available experiences are the SE Senior-level / Expert, and the rarest experiences are the Executive-level / Director

```python
fig = plt.figure(figsize=(25,8))
sns.countplot(df , x ='employee_residence' )
fig.suptitle('employees residence', fontsize=40, fontweight='bold')
```

```python
fig = plt.figure(figsize=(25,8))
sns.countplot(df , x ='company_location' )
fig.suptitle('Companies distribution in countries', fontsize=40, fontweight='bold')
```

From the previous two figures, we note that the data is biased towards companies and individuals from the United States

```python
sns.countplot(df , x ='company_size' )
plt.title('distribution of employees based on the size of the companies', fontsize=12, fontweight='bold')
```

Note that the samples are largely directed to medium-sized companies, With great neglect of small-sized companies

```python
sns.countplot(df , x ='employment_type' )

plt.title('ratios of employment types', fontsize=20, fontweight='bold')
```

Almost all employees work full time

```python
df['employee_works_in_his_country'] = [1 if i == j else 0 for i , j in zip(df['employee_residence'] , df['company_location'])]
plt.pie(df['employee_works_in_his_country'].value_counts() ,
        labels = ['works in his country' , 'works in a foreign country' ],
        autopct='%1.1f%%')
plt.title('number of employees working in there countries', fontsize=15, fontweight='bold')
```

Almost all employees work in there countries

```python
fig = plt.figure(figsize=(30,8))
sns.countplot(df , x ='job_title' )
plt.xticks(rotation=90)
fig.suptitle('job titles distribution', fontsize=40, fontweight='bold')
```

Most in demand jobs are Data scientist, Data Engineer, and Data Analyst

```python
sns.set(style="darkgrid")
sns.displot(data=df, 
            x='job_title',
            row="work_year" ,
            bins=20,
#              height = 15 ,
            aspect = 2.5
           )
plt.xticks(rotation=90)
```

The number of jobs in the Data Science field has dramatically doubled form 2020 to 2022
